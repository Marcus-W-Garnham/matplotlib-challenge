## Observations and Insights 

```python
%matplotlib notebook
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
# I use outer as I want all data to be in my dataframe and merge on only shared column Mouse ID.
mouse_combined = pd.merge(mouse_metadata, study_results, how="outer", on=["Mouse ID"])

# Display the data table for preview
mouse_combined
```

```python
# Checking the number of mice.

# This counts the number of unique values in Mouse ID using nunique which counts unique values in the column specified. 
Mouse_number = mouse_combined['Mouse ID'].nunique()

# Dipslay the number of mice
Mouse_number
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

# Can't use just Duplicated_Mouse_Row = mouse_combined[mouse_combined.duplicated()] ptherwise you only get one row
# This is beacuse it will only show rows that are an excat copy and only time that happens is at timepoint zero as everything
# starts at a control level for Tumor Volume (mm3) and Metastatic Sites and even if results are taken at the same time these
# two variables will vary as there is always some uncertainity/variance even if same person is recording results.

# For finding actual duplicates I need to specify in code which columns to check.
# The code below checks if the Mouse ID is duplicated which it will be as results are taken at several time points.
# Which is why we need to check if the Mouse ID has a duplicated timepoint.
Duplicated_Mouse_Row = mouse_combined[mouse_combined.duplicated(['Mouse ID', 'Timepoint'])]

# Note the table diplayed below is the full table as there are no **** rows x 8 columns as all data is in range nor is there 
# a ... line gap in middle of table to show we are jumping to the end.
Duplicated_Mouse_Row
```

```python
# Optional: Get all the data for the duplicate mouse ID.

# Since the only mouse that has duplicated data is g989 I only have to use one condition for this and the following code.
# If I had more that one I would probably use a for loop with a list of all duplicat mice.
duplicate = []
duplicate = mouse_combined.loc[mouse_combined['Mouse ID'] == 'g989']
duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

mouse_combined_cleaned = mouse_combined[mouse_combined['Mouse ID'] != 'g989']
mouse_combined_cleaned
```

```python
# Checking the number of mice in the clean DataFrame.

Mouse_number_clean = mouse_combined_cleaned['Mouse ID'].nunique()
Mouse_number_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

# MEAN
# Mean we need to grouby drug regimen and get the mean of the other columns with data in them
Regimen_Mean = mouse_combined_cleaned.groupby("Drug Regimen").mean()
# Change name of tumor volume so we know its the mean value.
Regimen_Mean = Regimen_Mean.rename(columns={'Tumor Volume (mm3)': 'Mean Tumor Volume (mm3)'})
# Need to remove unwanted and uneeded tables.
Regimen_Mean_cleaned = Regimen_Mean[['Mean Tumor Volume (mm3)']]

# MEDIAN
# Median we need to do the same as we did for mean just instead of .mean() we use .median()
Regimen_Median = mouse_combined_cleaned.groupby("Drug Regimen").median()
# Change name of tumor volume so we know its the median value.
Regimen_Median = Regimen_Median.rename(columns={'Tumor Volume (mm3)': 'Median Tumor Volume (mm3)'})
# Remove unwanted data
Regimen_Median_cleaned = Regimen_Median[['Median Tumor Volume (mm3)']]

# VARIANCE
# Variance we need to do the same as we did for mean just instead of .mean() we use .var()
Regimen_Var = mouse_combined_cleaned.groupby("Drug Regimen").var()
# Rename column since it needs to be clear its the variance of Tumor Volume
Regimen_Var = Regimen_Var.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume (mm3) Variance'})
# Remove unwanted data
Regimen_Var_cleaned = Regimen_Var[['Tumor Volume (mm3) Variance']]

# STANDARD DEVIATION
# Standard Deviation we need to do the same as we did for mean just instead of .mean() we use .std()
Regimen_Stan_Dev = mouse_combined_cleaned.groupby("Drug Regimen").std()
# Rename column since it needs to be clear its the Standard Deviation of Tumor Volume
Regimen_Stan_Dev = Regimen_Stan_Dev.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume (mm3) Standard Deviation'})
# Remove unwanted data
Regimen_Stan_Dev_cleaned = Regimen_Stan_Dev[['Tumor Volume (mm3) Standard Deviation']]

# SEM (standard error of the mean)
# SEM we need to do the same as we did for mean just instead of .mean() we use .sem()
Regimen_SEM = mouse_combined_cleaned.groupby("Drug Regimen").sem()
Regimen_SEM = Regimen_SEM.rename(columns={'Tumor Volume (mm3)': 'SEM Tumor Volume (mm3)'})
Regimen_SEM_cleaned = Regimen_SEM[['SEM Tumor Volume (mm3)']]
Regimen_SEM_cleaned

# Assemble the resulting series into a single summary dataframe.
Regimen_sumarry_1 = pd.merge(Regimen_Mean_cleaned, Regimen_Median_cleaned, how="outer", on="Drug Regimen")
Regimen_sumarry_2 = pd.merge(Regimen_sumarry_1, Regimen_Var_cleaned, how="outer", on="Drug Regimen")
Regimen_sumarry_3 = pd.merge(Regimen_sumarry_2, Regimen_Stan_Dev_cleaned, how="outer", on="Drug Regimen")
Regimen_sumarry_final = pd.merge(Regimen_sumarry_3, Regimen_SEM_cleaned, how="outer", on="Drug Regimen")
Regimen_sumarry_final
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

#remove all data on about mice not on Capomulin
mice_on_Capomulin = mouse_combined_cleaned[mouse_combined_cleaned['Drug Regimen']=='Capomulin']

#from this data I will pick mouse ID i557 to plot

#remove all data not on mouse i557
Mouse_i557 = mice_on_Capomulin[mice_on_Capomulin['Mouse ID']=='i557']

# setup figure size then put in the plot information then make layout tight
plt.figure(figsize=(10,10))
plt.plot(Mouse_i557['Timepoint'], Mouse_i557['Tumor Volume (mm3)'])
plt.tight_layout()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
Mean_Tumor_Capomulin = mice_on_Capomulin.groupby('Mouse ID').mean('Tumor Volume (mm3)')

plt.figure(figsize=(10,10))
plt.scatter(Mean_Tumor_Capomulin['Tumor Volume (mm3)'], Mean_Tumor_Capomulin['Weight (g)'])
plt.xlabel('Tumor Volume')
plt.ylabel('Mouse Weight')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
plt.figure(figsize=(10,10))
plt.scatter(Mean_Tumor_Capomulin['Tumor Volume (mm3)'], Mean_Tumor_Capomulin['Weight (g)'])
plt.xlabel('Tumor Volume')
plt.ylabel('Mouse Weight')

x_values = Mean_Tumor_Capomulin['Tumor Volume (mm3)']
y_values = Mean_Tumor_Capomulin['Weight (g)']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, "r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
```
